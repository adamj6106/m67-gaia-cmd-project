```python
import pandas as pd

df = pd.read_csv("M67-research-result.csv")

print(df.head())
print(df.columns)
```

```python
print(df.head())
print(df.columns)
print(df.shape)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
plt.scatter(df["bp_rp"], df["phot_g_mean_mag"], s=2)
plt.gca().invert_yaxis()

plt.xlabel("Color Index (BP - RP)")
plt.ylabel("Gaia G Magnitude")
plt.title("M67 Color-Magnitude Diagram")

plt.show()
```

```python
filtered = df[
    (df["parallax"] > 0.9) &
    (df["parallax"] < 1.4) &
    (df["pmra"] > -14) &
    (df["pmra"] < -8) &
    (df["pmdec"] > -7) &
    (df["pmdec"] < -2)
]

print(filtered.shape)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(filtered["bp_rp"], filtered["phot_g_mean_mag"], s=6, alpha=0.7)

plt.gca().invert_yaxis()
plt.xlabel("Color Index (BP - RP)")
plt.ylabel("Gaia G Magnitude")
plt.title("M67 Color-Magnitude Diagram - Filtered Likely Members")

plt.show()
```
